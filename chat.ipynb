Blocking	Non-blocking
No await	Requires await
One request at a time	Many requests concurrently
Simple scripts	FastAPI, AI backends

````python
llm = ChatOllama(
    model="gemma3:1b",
    temperature=0,
    num_predict=100
)
for chunk in llm.stream(
    "Tell me a joke"
):  
    # try:
        # print(type(chunk))
        # print(chunk)
        if (not chunk.response_metadata.get('created_at')):
            print(chunk.content, end="")
        else:
            metadata = chunk.response_metadata
            usage = chunk.usage_metadata
            print("\n \n", chunk)
            seconds = metadata.get('total_duration')/ 1000000000 
            print(f"Timetaken: {seconds} seconds, reason: {metadata.get('done_reason')}")
            print(f"Output tokens: {usage.get('output_tokens')}, input tokens: {usage.get('input_tokens')}")
            break
        
    
````

````python
 

llm = ChatOllama(
    model="gemma4:e2b",
    temperature=0.7,
    num_predict=50,
    reasoning=False
)
async for chunk in llm.astream(
    "Tell me a joke"
):  
    # try:
        # print(type(chunk))
        # print(chunk)
        if ((chunk.additional_kwargs.get('reasoning_content'))):
            print(chunk.additional_kwargs.get('reasoning_content'), end="")
        elif (not chunk.response_metadata.get('created_at')):
            print(chunk.content, end="")
            # print(chunk.response_metadata)
        else:
            metadata = chunk.response_metadata
            usage = chunk.usage_metadata
            print("\n \n", chunk)
            seconds = metadata.get('total_duration')/ 1000000000 
            print(f"Timetaken: {seconds} seconds, reason: {metadata.get('done_reason')}")
            print(f"Output tokens: {usage.get('output_tokens')}, input tokens: {usage.get('input_tokens')}")
            break
````

*** Note: no change in  format for llm.stream() and llm.astream() in output wise. The difference is the code execution 
where stream blocks the eventloop thread till completion while llm.astream pauses and resumes without blocking the event loop thread


* 1. **invoke**
  * The Use: Simple, straightforward, one-shot execution.
  * When to use it:In simple scripts, background cron jobs, or data pipelines where you do not care about blocking execution.
  * When the output is short and you need the full data before moving to the next line of code (e.g., classifying text into categories, parsing variables).
  * **Example:** Running a script that classifies 100 customer reviews as "Positive" or "Negative" one by one.

* 2. **ainvoke**
  * The Use: High-performance, concurrent, one-shot execution.
  * When to use it:Inside modern web frameworks like FastAPI or Sanic.When you need to make multiple LLM calls at the exact same time without waiting for the first one to finish before starting the second.
  * **Example:** A user logs in, and you concurrently run 3 different prompts in the background: one to summarize their history, one to check their profile context, and one to generate a greeting. ainvoke lets them run simultaneously.

* 3. **stream**
* The Use: Instant visual feedback in simple, synchronous programs.
  * When to use it:Inside command-line interfaces (CLI tools) or frameworks like Streamlit.When the LLM output is long (like an essay or a code block) and you don't want the user staring at a blank screen for 10 seconds while the model generates the full answer
  * **Example:** A terminal-based chatbot helper where text drops onto the screen word-by-word.

* 4. **astream**
  * The Use: The gold standard for modern, production-grade web applications.
  * When to use it:When building a ChatGPT-like web interface where words stream in real-time, but hundreds of users are hitting the server at the exact same time.
  * If you used standard stream with 10 users simultaneously, the server would lock up and freeze users out. astream allows the server's event loop to juggle all 100 users flawlessly while streaming text to everyone.
  * **Example:** A production chat app frontend connected to a FastAPI backend using Server-Sent Events (SSE) or WebSockets.
